```python
import numpy as np
import math
# Add the missing attribute back manually
if not hasattr(np, "product"):
    np.product = np.prod
np.math = math
import matplotlib.pyplot as plt
import scipy.constants as cs
import time

import rydcalc

%load_ext memory_profiler
```

```python
Yb171 = rydcalc.Ytterbium171(cpp_numerov=True,use_db=False)
Rb = rydcalc.Rubidium87(cpp_numerov=True,use_db=False)

includefn = lambda p,p0: True if ((np.abs(p0.energy_Hz-p.energy_Hz)<5e9 ) and ((-1)**(p.s1.channels[0].l + p.s2.channels[0].l)==(-1)**(p0.s1.channels[0].l + p0.s2.channels[0].l))) else False
opts = {'dn': 2,'dl': 2,'dm': 2,'dipole_allowed': False, 'pair_include_fn': lambda p,p0: np.abs(p0.energy_Hz-p.energy_Hz)<1e9}

search = rydcalc.rydberg_pair_search(Yb171, Rb, [50, 70], [50, 70], [10, 10], opts=opts, save_file='first_search.h5')
```

```python
search.generate_search_space(100, 1)
print(search.search_space_a1a2)
print(search.search_space_a1a1)
print(search.search_space_a2a2)
assert(len(search.search_space_a1a2) == len(search.search_space_a1a1) * len(search.search_space_a2a2))
```

```python
if __name__ == '__main__':
    search.run_search()
```
